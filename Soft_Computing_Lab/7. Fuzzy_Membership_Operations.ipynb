```python
# Functions to use further

import numpy as np
import matplotlib.pyplot as plt

def setRange():
    print("Enter the ranges ->")
    f_range = float(input("Enter the lower limit: "))
    l_range = float(input("Enter the upper limit"))
    num_divisions = int(input("Enter the number of divisions: "))
    x_values = np.linspace(f_range, l_range, num_divisions)

    
def plotting(x_values, membership_degrees, x_label, y_label, title):
    plt.plot(x_values, membership_degrees, color='blue')
    plt.xlabel(x_label)
    plt.ylabel(y_label)
    plt.title(title)
    plt.grid(True)
    plt.show()
```

```python
# Triangular membership function

def triangular(x, a, b, c):
    return max(min((x - a) / (b - a), (c - x) / (c - b)), 0)

setRange()

x1_value = float(input("Enter value of x1: "))
x2_value = float(input("Enter value of x2: "))
x3_value = float(input("Enter value of x3: "))

membership_degrees = [triangular(x, x1_value, x2_value, x3_value) for x in x_values]

plotting(x_values, membership_degrees, 'x', 'Degree of Membership', 'Triangular Membership Function')
```

```python
# Trapezoidal Membership Function

def trapezoidal(x, a, b, c, d):
    return max(min((x - a) / (b - a), 1, (d - x) / (d - c)), 0)

setRange()

x1_value = float(input("Enter value of x1: "))
x2_value = float(input("Enter value of x2: "))
x3_value = float(input("Enter value of x3: "))
x4_value = float(input("Enter value of x4: "))

membership_degrees = [trapezoidal(x, x1_value, x2_value, x3_value, x4_value) for x in x_values]

plotting(x_values, membership_degrees, 'x', 'Degree of Membership', 'Trapezoidal Membership Function')
```

```python
# Sigmoid Function

def sigmoid(x):
    return 1.0 / (1.0 + np.exp(-x))

setRange()

y_values = sigmoid(x_values)

plotting(x_values, y_values, 'x',  'Sigmoid(x)', 'Sigmoid Function')
```

```python
# Generalized Bell Function

def gbellmf(x, center, width, shape):
    return 1 / (1 + np.abs((x - center) / width) ** (2 * shape))

setRange()

center_value = int(input("Enter the center value: "))
width_value = int(input("Enter the width of the curve: "))
shape_value = int(input("Enter the shape value of the curve: "))

y_values = gbellmf(x_values, center_value, width_value, shape_value)

plotting(x_values, y_values, 'x',  'Degree of Membership', 'Generalized Bell Function')
```

```python
# Gaussian Function

def pdf(x): 
    mean = np.mean(x) 
    std = np.std(x) 
    y_out = 1/(std * np.sqrt(2 * np.pi)) * np.exp( - (x - mean)**2 / (2 * std**2)) 
    return y_out 

setRange()

y_values = pdf(x_values) 

plotting(x_values, y_values, 'x',  'Degree of Membership', 'Gaussian Function')
```
